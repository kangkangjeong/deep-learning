```python
#다중선형회귀
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1=np.array([2,4,6,8])
x2=np.array([0,4,2,3])
y=np.array([81,93,91,97])

```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter3D(x1,x2,y)
plt.show()
```

```python
a1,a2,b=(0,0,0)
lr=0.01
epochs=2001
n=len(x1)
#y_pred=a1*x1+a2*x2+b


#경사 하강법을 시작
for i in range(epochs):
    y_pred=a1*x1+a2*x2+b
    error=y-y_pred
    
    a1=a1-lr*((2/n)*sum(-x1*(error)))
    a2=a2-lr*((2/n)*sum(-x2*(error)))
    b=b-lr*((2/n)*sum(-(error)))
    
    if i%100==0:
        print('epochs:{},기울기1:{:3.2f},기울기2:{:3.2f},절편:{:3.2f}'.format(i,a1,a2,b))
#실제 예측된 최종 점수 출력
y_pred=a1*x1+a2*x2+b
print('실제점수:',y)
print('예측점수:',y_pred)
print('예측점수:',a1*5+a2*3+b)
```

```python
#텐서플로우와 케라스를 활용한 선형 회귀 분석
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten

model=Sequential()
model.add(Dense(1,input_dim=1,activation='linear')) # layer 1
model.summary()
model.compile(optimizer='sgd',loss='mse') #모델 파라미터 설정

model.fit(x1,y,epochs=2000) # 모델학습
```

```python

plt.scatter(x1,y)
plt.plot(x1,model.predict(x1),'r')
plt.show()
```

```python
# 공부 시간이 5시간인 학생의 점수 예측
model.predict([5])
```

```python
#코딩으로 XOR 문제 해결하기

w11=np.array([-2,2]) #첫번째 레이어의 첫번쨰 가중값
w12=np.array([2,2])
w2 = np.array([1,1])#첫번째 레이어의 두번째 가중치
b1=np.array([1,1]) #두번째 레이어의 가중치 #bias1
b1=3
b2=-1
b3=-1
```

```python
#퍼셉트론 함수 정의
def MLP(x,w,b):
    y=np.sum(w*x)+b
    if y<=0:
        return 0
    else:
        return 1
#NAND 게이트
def NAND(x1,x2):
    return MLP(np.array([x1,x2]),w11,b1)

#OR 게이트
def OR(x1,x2):
    return MLP(np.array([x1,x2]),w12,b2)

#AND 게이트
def AND(x1,x2):
    return MLP(np.array([x1,x2]),w2,b3)

#XOR 게이트
def XOR(x1,x2):
    return AND(NAND(x1,x2),OR(x1,x2))
```

```python
for x in [(0,0),(0,1),(1,0),(1,1)]:
    y=XOR(x[0],x[1])
    print('입력 값:',str(x),'출력 깂:',str(y))
``
```
